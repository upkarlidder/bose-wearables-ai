```python
import pandas as pd
from datetime import datetime
import numpy as np

sensordata = pd.read_csv("sensor-data.csv")

accel = pd.DataFrame(sensordata, columns=['timestamp','x','y','z']).dropna() #drop NaNs
attitude = pd.DataFrame(sensordata, columns=['timestamp','pitch','roll','yaw','pitchRadian','rollRadian','yawRadian']).dropna() #drop NaNs
accel = accel.set_index('timestamp')
attitude = attitude.set_index('timestamp')

result = accel.join(attitude)

x = pd.DataFrame(result['x'])
y = pd.DataFrame(result['y'])
z = pd.DataFrame(result['z'])


pitch = pd.DataFrame(result['pitch'])
roll = pd.DataFrame(result['roll'])
yaw = pd.DataFrame(result['yaw'])

```

```python
print(x.plot())
print(x.describe())
print(y.plot())
print(y.describe())
print(z.plot())
print(z.describe())
```

```python
print(pitch.plot())
print(roll.plot())
print(yaw.plot())
print(yaw.describe())
```
